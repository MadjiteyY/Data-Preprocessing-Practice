#### Getting the tweet image prediction data (image_predictions.tsv)

```python
#using the Requests library to download the tweet image prediction
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
r = requests.get(url)

#save response (r) to a file (image_predictions.tsv)
with open ('image_predictions.tsv', mode = 'wb') as file:
    file.write(r.content)
    
#read file into pandas dataframe
tweet_image_pred = pd.read_csv('image_predictions.tsv', sep = '\t')
```

#### Getting additional data via the Twitter API 

```python
#Using the Tweepy library to query additional data via the Twitter API
#Setup: getting started with the tweepy libray
#my access and consumer details are not provided here for security reasons
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
#the parser here directly gets the json_data from the status
api = tweepy.API(auth_handler=auth, parser=tweepy.parsers.JSONParser(), wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

#getting the statuses and saving their json data into a file
failed_ids = {}
#to open the .txt file
with open ('tweet_json.txt', mode = 'w') as file:
    #the for loop gets the status of each tweet_id
    for tweet_id in twitter_archive.tweet_id.values:
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print(str(tweet_id) + ":" + "success")
            #saves the data in the file
            json.dump(tweet, file)
            #sets each data on a line
            file.write('\n')
        except Exception as e:
            #gets ids with no statuses or ids with tweepy.error
            failed_ids[tweet_id] = e
            pass
print(failed_ids)
```

```python
#read the .txt file, line by line, to create a pandas DataFrame
tweet_list = []
with open('tweet_json.txt', mode = 'r') as f:
    # to read file line by line and load data
    for line in f:
        tweet_data = json.loads(line)
        #to create keys, values pair or dictionary
        tweet_list.append({"tweet_id": tweet_data["id_str"],
                          "retweet_count": tweet_data["retweet_count"],
                          "favorite_count": tweet_data["favorite_count"],
                          "created_at": tweet_data["created_at"]})
        #to create a pandas dataframe
        tweet_data = pd.DataFrame(tweet_list, columns = ["tweet_id", "retweet_count", "favorite_count", "created_at"])
```

3. not all the data are dog ratings and some are retweets.
  eg. 
      - index 19 =RT @dog_rates: This is Canela,... (Retweet)
      - index 23 = This is Canela,... (original tweet)
      - index 25 is not a dog.
4. inaccurate dog names: dog names entered as (such, quite, a, an, the, not, one, by)
5. inaccurate rating values (eg. 27 entered in rating_numerator instead of 11.27 as given in text)
   ii. some rating_numerator values within text are of float type

   which is a string value instead of an empty string to depict missing values

***`Tweet_image_pred`***
1. tweet_id is of integer datatype
2. inconsistency in object name ie. mix of lower and upper case
3. some images are not dogs
4. underscores in columns p1, p2, p3

***`Tweet_data`***

1. created_at is of object datatype (to be dropped)
